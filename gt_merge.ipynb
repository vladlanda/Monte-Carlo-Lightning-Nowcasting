```python
import pandas as pd
```

```python

def merge_excel_files(file_list):
    merged_df = None
    for file in file_list:
        if file.endswith('.xlsx'):df = pd.read_excel(file)
        elif file.endswith('.csv'):df = pd.read_csv(file)
        else:
            raise ValueError(f"Unsupported file format: {file}")
        # Only keep the specified columns
        df = df[['UTC', 'lat', 'lon']]
        
        if merged_df is None:
            merged_df = df
        else:
            # Concatenate along rows (axis=0)
            merged_df = pd.concat([merged_df, df], axis=0)
    return merged_df

# file_list = ['ILDN 2022-2023 season.xlsx',
#              'ILDN 2023-2024 season.xlsx', 
#              'ENTLN 2023-2024 season.csv', 
#              'ENTLN 2022-2023 season.xlsx', 
#              'WWLN 2023-2024 season.xlsx']  # replace with your file list

file_list = [
             'ILDN 2025-2027 season-cases-only.xlsx', 
             'ENTLN 2025-2026 season.xlsx', 
             ]  # replace with your file list

merged_df = merge_excel_files(file_list)

# Save the merged DataFrame as an Excel file
merged_df.to_csv('GT 2025-2027 season.csv', index=False)
# merged_df.to_excel('GT 2023-2024 season.xlsx', index=False,sheet_name='Sheet1')


```
